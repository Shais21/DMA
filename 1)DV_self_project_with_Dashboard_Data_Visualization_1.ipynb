````python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Load the dataset
data_path = 'mtcars.csv'  # Ensure this file is in the same directory as the script
data = pd.read_csv(data_path)

# Initialize the Dash app
app = dash.Dash(__name__)
app.title = 'MTCars Dashboard'

# Layout of the app
app.layout = html.Div([
    html.H1('MTCars Dashboard', style={'textAlign': 'center'}),

    html.Div([
        html.Label('Select X-axis:'),
        dcc.Dropdown(
            id='x-axis',
            options=[{'label': col, 'value': col} for col in data.columns],
            value='mpg'
        ),

        html.Label('Select Y-axis:'),
        dcc.Dropdown(
            id='y-axis',
            options=[{'label': col, 'value': col} for col in data.columns],
            value='hp'
        ),

        html.Label('Select Color By:'),
        dcc.Dropdown(
            id='color-by',
            options=[{'label': col, 'value': col} for col in data.columns],
            value='cyl'
        ),
    ], style={'width': '30%', 'display': 'inline-block', 'verticalAlign': 'top'}),

    dcc.Graph(id='scatter-plot', style={'width': '65%', 'display': 'inline-block'}),

    html.Div([
        html.Label('Select Column for Histogram:'),
        dcc.Dropdown(
            id='histogram-column',
            options=[{'label': col, 'value': col} for col in data.columns],
            value='mpg'
        ),
    ], style={'width': '30%', 'display': 'inline-block', 'verticalAlign': 'top'}),

    dcc.Graph(id='histogram', style={'width': '65%', 'display': 'inline-block'}),
])

# Callbacks to update the plots
@app.callback(
    Output('scatter-plot', 'figure'),
    [Input('x-axis', 'value'),
     Input('y-axis', 'value'),
     Input('color-by', 'value')]
)
def update_scatter(x_axis, y_axis, color_by):
    fig = px.scatter(
        data, x=x_axis, y=y_axis, color=color_by,
        title=f'Scatter Plot of {y_axis} vs {x_axis}',
        labels={x_axis: x_axis, y_axis: y_axis, color_by: color_by}
    )
    return fig

@app.callback(
    Output('histogram', 'figure'),
    [Input('histogram-column', 'value')]
)
def update_histogram(column):
    fig = px.histogram(
        data, x=column, nbins=10,
        title=f'Histogram of {column}',
        labels={column: column}
    )
    return fig

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
````
